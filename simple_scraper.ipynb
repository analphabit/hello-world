```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#the whole scraper in one cell

data = []

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")

films = soup.find('tbody', class_='lister-list').findAll('tr') 
for film in films:
    name = film.find('td', class_='titleColumn').find('a').text
    link = "https://www.imdb.com" + film.find('td', class_='titleColumn').find('a').get('href')
    year = film.find('td', class_='titleColumn').find('span', class_='secondaryInfo').text
    rate = film.find('td',class_ = "ratingColumn").text.strip()
    
    print(name, link, year, rate)
    data.append([name, link, year, rate])   
```

```python
#write our date to csv
header = ['name', 'link', 'year', 'rate']
df = pd.DataFrame(data, columns=header)
df.head()
```

```python
df.to_csv("250_films.csv")
```
